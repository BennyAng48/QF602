### down and in/out Call Option Deal Specification

```python
k = 100
h = 90
T = 1
q = 0.0
r = 0.0
vol = 0.20
Spots = np.linspace(70,120,200)
```

```python
dic = [mybo.BarrierOptionPrice(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_IN_CALL) for S0 in Spots]
doc = [mybo.BarrierOptionPrice(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_OUT_CALL) for S0 in Spots]
eurocall = [i + o for i, o in zip(dic, doc)]

output = np.array([Spots, dic, doc, eurocall])
output = np.transpose(output)
df = pd.DataFrame(output, columns =['S_0', 'DIC price', 'DOC price', 'Call price']) 
df = df.set_index('S_0')
df.plot(grid=True, figsize=(7, 5))
plt.title("K = 100, H = 90, vol = 0.2, T=1, r=0, q=0")
plt.xlabel("Stock Price")
plt.ylabel("Option Price")
#plt.savefig("L5_DI_Price.png")
plt.show()
```

```python
dic_delta = [mybo.BarrierOptionDelta_fd(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_IN_CALL) for S0 in Spots]
doc_delta = [mybo.BarrierOptionDelta_fd(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_OUT_CALL) for S0 in Spots]
eurocall_delta = [i + o for i, o in zip(dic_delta, doc_delta)]

output = np.array([Spots, dic_delta, doc_delta, eurocall_delta])
output = np.transpose(output)
df = pd.DataFrame(output, columns =['S_0', 'DIC delta', 'DOC delta', 'Call delta']) 
df = df.set_index('S_0')
df.plot(grid=True, figsize=(7, 5))
plt.title("K = 100, H = 90, vol = 0.2, T=1, r=0, q=0")
plt.xlabel("Stock Price")
plt.ylabel("Option Delta")
#plt.savefig("L5_DI_Delta.png")
plt.show()
```

```python
dic_gamma = [mybo.BarrierOptionGamma_fd(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_IN_CALL) for S0 in Spots]
doc_gamma = [mybo.BarrierOptionGamma_fd(S0, r, q, k, h, vol, T, mybo.BarrierTypes.DOWN_AND_OUT_CALL) for S0 in Spots]
eurocall_gamma = [i + o for i, o in zip(dic_gamma, doc_gamma)]

output = np.array([Spots, dic_gamma, doc_gamma, eurocall_gamma])
output = np.transpose(output)
df = pd.DataFrame(output, columns =['S_0', 'DIC gamma', 'DOC gamma', 'Call gamma']) 
df = df.set_index('S_0')
df.plot(grid=True, figsize=(7, 5))
plt.title("K = 100, H = 90, vol = 0.2, T=1, r=0, q=0")
plt.xlabel("Stock Price")
plt.ylabel("Option Gamma")
#plt.savefig("L5_DI_Gamma.png")
plt.show()
```

### Barrier over hedge

```python
k = 100
h = 130
hadj = 132
T = 0.5
q = 0.0
r = 0.0
vol = 0.10
Spots = np.linspace(80,140,200)

uoc = [mybo.BarrierOptionPrice(S0, r, q, k, h, vol, T, mybo.BarrierTypes.UP_AND_OUT_CALL) for S0 in Spots]
uoc_adj = [mybo.BarrierOptionPrice(S0, r, q, k, hadj, vol, T, mybo.BarrierTypes.UP_AND_OUT_CALL) for S0 in Spots]
value_cushion = [adj - norm for norm, adj in zip(uoc, uoc_adj)]

output = np.array([Spots, uoc, uoc_adj, value_cushion])
output = np.transpose(output)
df = pd.DataFrame(output, columns =['S_0', 'UOC', 'UOC adjust', 'value cushion']) 
df = df.set_index('S_0')
df.plot(grid=True, figsize=(7, 5))
plt.title("K = 100, H = 130, vol = 0.1, T=0.5, r=0, q=0")
plt.xlabel("Stock Price")
plt.ylabel("Option Price")
#plt.savefig("L5_Barrier_over_hedge.png")
plt.show()
```

```python
uoc_delta = [mybo.BarrierOptionDelta_fd(S0, r, q, k, h, vol, T, mybo.BarrierTypes.UP_AND_OUT_CALL) for S0 in Spots]
uoc_delta_adj = [mybo.BarrierOptionDelta_fd(S0, r, q, k, hadj, vol, T, mybo.BarrierTypes.UP_AND_OUT_CALL) for S0 in Spots]

output = np.array([Spots, uoc_delta, uoc_delta_adj])
output = np.transpose(output)
df = pd.DataFrame(output, columns =['S_0', 'UOC delta', 'UOC adj delta']) 
df = df.set_index('S_0')
df.plot(grid=True, figsize=(7, 5))
plt.title("K = 100, H = 130, vol = 0.1, T=0.5, r=0, q=0")
plt.xlabel("Stock Price")
plt.ylabel("Option Delta")
#plt.savefig("L5_Barrier_over_hedge_delta.png")
plt.show()
```
