```python
import numpy as np
import matplotlib.pyplot as plt
from financepy.models.black import *
from financepy.utils.global_types import *
```

```python
K = 100
q = 0.00
r = 0.00
ivol = 0.20
model = Black(ivol)
callOrPut = OptionTypes.EUROPEAN_CALL
T1m = 1.0/12; T6m = 1/2; T1y = 1
DF1m = np.exp(-r*T1m); DF6m = np.exp(-r*T6m); DF1y = np.exp(-r*T1y)

Spots = np.linspace(50,150,100)
forward1m = [S*np.exp(-q*T1m)/DF1m for S in Spots]
forward6m = [S*np.exp(-q*T6m)/DF6m for S in Spots]
forward1y = [S*np.exp(-q*T1y)/DF1y for S in Spots]
```

### Delta profiles for different maturities

```python
delta1m = [model.delta(f, K, T1m, DF1m, callOrPut) for f in forward1m]
delta6m = [model.delta(f, K, T6m, DF6m, callOrPut) for f in forward6m]
delta1y = [model.delta(f, K, T1y, DF1y, callOrPut) for f in forward1y]

plt.plot(Spots, delta1m, 'b', Spots, delta6m, 'g', Spots, delta1y, 'r')
plt.grid(linestyle='-')
plt.xlabel("Stock Price")
plt.ylabel("Option Delta")
```

### Gamma profiles for different maturities

```python
gamma1m = [model.gamma(f, K, T1m, DF1m, callOrPut) for f in forward1m]
gamma6m = [model.gamma(f, K, T6m, DF6m, callOrPut) for f in forward6m]
gamma1y = [model.gamma(f, K, T1y, DF1y, callOrPut) for f in forward1y]

plt.plot(Spots, gamma1m, 'b', Spots, gamma6m, 'g', Spots, gamma1y, 'r')
plt.grid(linestyle='-')
plt.xlabel("Stock Price")
plt.ylabel("Option Gamma")
```

```python
dollargamma1m = [model.gamma(f, K, T1m, DF1m, callOrPut)*s/100 for f,s in zip(forward1m, Spots)]
dollargamma6m = [model.gamma(f, K, T6m, DF6m, callOrPut)*s/100 for f,s in zip(forward6m, Spots)]
dollargamma1y = [model.gamma(f, K, T1y, DF1y, callOrPut)*s/100 for f,s in zip(forward1y, Spots)]

plt.plot(Spots, dollargamma1m, 'b', Spots, dollargamma6m, 'g', Spots, dollargamma1y, 'r')
plt.grid(linestyle='-')
plt.xlabel("Stock Price")
plt.ylabel("dollar Option Gamma")
```
